```python
import pandas as pd
import matplotlib.pyplot as plt
import os

print(os.getcwd())

# Check if file exists
file_path = '../scraper/binance_data_merged/klines/30m/merged.csv'
if not os.path.exists(file_path):
    raise FileNotFoundError(f"CSV file not found at path: {file_path}")

# Read the CSV data
df = pd.read_csv(file_path)

# Create the plot
plt.figure(figsize=(15,8))

# Plot candlestick chart
plt.plot(pd.to_datetime(df['open_time'], unit='ms'), df['close'], label='Close Price')

# Add volume bars at the bottom
ax2 = plt.gca().twinx()
ax2.fill_between(pd.to_datetime(df['open_time'], unit='ms'), df['volume'], alpha=0.3, color='gray', label='Volume')

# Customize the plot
plt.title('Bitcoin Price and Volume Over Time')
plt.xlabel('Time')
plt.ylabel('Price (USD)')
ax2.set_ylabel('Volume')

# Add legend
lines1, labels1 = plt.gca().get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines1 + lines2, labels1 + labels2, loc='upper left')

plt.grid(True)
plt.show()
```
